```python
#calculate factorial of a number
num=int(input("enter a number:"))
fact=1
for i in range(1,num+1):
    fact=fact*i
    print("factorial of",num,"is",fact)
```

```python
#sum of 1st n natural number
n=int(input("Enter a number:"))
sum=0
i=1
while i <=n:
   sum=sum+i
   i=i+1
   print("sum=",sum)
```

```python
#reverse a number
num=int(input("enter a number:"))
rev=0
while num>0:
    digit=num%10
    rev=rev*10+digit
    num=num//10
    print("Reversed number is",rev)
```

```python
# count digits in a number
num=int(input("enter a number:"))
count=0
if num==0:
    count=1
else:
    while num!=0:
        num=num//10
        count+=1
        print("number of digits",count)
```

```python
#stop printing numbers when a specific number is found
n=int(input("enter the limit:"))
stop=int(input("enter the number to stop at:"))
for i in range(1,n+1):
    if i==stop:
        break
        print(i)
```

```python
#search for an element in a list and stop when found
nums=[10,20,30,40,50]
key=int(input("enter number to search:"))
for i in nums:
    if i==key:
        print("element found",i)
        break
    else:
        print("element not found")
```
